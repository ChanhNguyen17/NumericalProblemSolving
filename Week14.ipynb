```python
def f(x):
    return -26 + 85*x - 91*x**2 + 44*x**3 - 8*x**4 + x**5
def fprime(x):
    return 85 - 2*91*x + 3*44*x**2 - 4*8*x**3 + 5*x**4
```

```python
x = 0.0
for i in range(10):
    x = x - f(x)/fprime(x)
    print(x, f(x), fprime(x))
```

```python
def fsecant(x, dx = 1e-6):
    return (f(x + dx) - f(x))/dx
```

```python
x = 0.0
for i in range(10):
    x = x - f(x)/fsecant(x, 0.001)
    print(x, f(x))
```
