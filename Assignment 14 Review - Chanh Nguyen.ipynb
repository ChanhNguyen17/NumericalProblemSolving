## Problem 2
Determine the roots of ${\large f(x) = -13 - 20x + 19x^2 - 3x^3}$  
(a) Graphically  
(b) Using simple fixed-point iteration  
(c) Finding the roots of the polynomial  

```python
x = np.linspace(-1, 5, 10e6)
def f(x):
    return -13 - 20*x + 19*x**2 - 3*x**3

y = f(x)
plot(x, y)
grid()
axhline(0, color = 'r')
```

###### (a) Graphically

```python
fig, ax = subplots(1, 3, figsize = (17, 5))

sca(ax[0])
plot(x, y)
xlim([-0.44688875, -0.44688873])
ylim([-0.000001, 0.000001])
axhline(0, color = 'r')
grid()

sca(ax[1])
plot(x, y)
xlim([2.04989415, 2.04989417])
ylim([-0.000001, 0.000001])
axhline(0, color = 'r')
grid()

sca(ax[2])
plot(x, y)
xlim([4.73032790, 4.73032792])
ylim([-0.000001, 0.000001])
axhline(0, color = 'r')
grid()
```

###### Answer: By using graphically method, the root is x1 = -0.44688874, x2 = 2.04989416, x3 = 4.73032791

```python
def fx(x):
    return (-13 - 20*x + 19*x**2)/(3*x**2)

def simpleFixedPointIteration(x0, fx, eS, maxIter):
    ite = 0
    xr = x0
    ea = 1.0
    while(ite<maxIter):
        ite+=1
        xrold = xr
        xr = fx(xr)
        ea = abs((xr-xrold)/xr)*100
        if(ea<eS):
            return xr, ite
    return xr, ite
```

###### (b) Using simple fixed-point iteration

```python
x, n = simpleFixedPointIteration(1.0, fx, 10e-2, 100)
print('x, n =', (x, n))
```

###### (c) Finding the roots of the polynomial

```python
x = polyroots([-13., -20, 19, -3])
print('x =', x)
```

## Problem 4  
An investigator has reported the data tabulated below for an experiment
to determine the growth rate of bacteria k (per day), as a function
of oxygen concentration c (mg/L). It is known that such data can be
modeled by the following equation:  
$\hspace{10mm} {\large k = \frac{k_{max}c^2}{c_s+c^2}}$

where cs and kmax are parameters. Use a transformation to linearize
this equation. Then use linear reqression to estimate cs and kmax and
predict the growth rate at c = 2 (mg/L).  
<table style="width:100%">
    <tr>
        <th>c</th>
        <td>0.5</td>
        <td>0.8</td>
        <td>1.5</td>
        <td>2.5</td>
        <td>4</td>

    </tr>
    <tr>
        <th>k</th>
        <td>1.1</td>
        <td>2.4</td>
        <td>5.3</td>
        <td>7.6</td>
        <td>8.9</td>

    </tr>
</table>  

```python
c = array([0.5, 0.8, 1.5, 2.5, 4])
k = array([1.1, 2.4, 5.3, 7.6, 8.9])

fig, ax = subplots(1, 2, figsize = (12, 5))

# The data has a linear relationship between log(c) and k. Use linregress for log(c) and k
clog = log(c)
r = linregress(k, clog)
a = r.slope
b = r.intercept

# Create equation based on linregress result
k2 = linspace(1, 9)
c2log = a*k2 + b

# Plot graph x and log(y)
sca(ax[1])
title('log(c) and k')
plot(clog, k, 'o')
plot(c2log, k2)
grid()

# Plot graph x and y
sca(ax[0])
c2 = exp(c2log)
plot(c, k, 'o')
plot(c2, k2)
title('c and k')
grid()

'The model equation: c = exp(a*k + b) with a = '+str(r.slope)+' and b = '+str(r.intercept)
```

###### Use linear reqression to estimate cs and kmax
${\large k = \frac{k_{max}c^2}{c_s+c^2} => \frac{1}{k} = \frac{c_s}{k_{max}} . \frac{1}{c^2} + \frac{1}{k_{max}}}$

```python
r = linregress(1/c**2, 1/k)
kmax = 1/r.intercept
cs = r.slope*kmax
print('cs = ', cs)
print('kmax = ',kmax)
```

###### Predict the growth rate at c = 2 (mg/L)

```python
c = 2
k = (kmax*c**2)/(cs + c**2)
k
```
