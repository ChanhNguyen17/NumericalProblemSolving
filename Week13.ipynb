```python
def trapezd(yf,x,n):
    step = (x[len(x)-1]-x[0])/n
    res = yf(x[0])+yf(x[len(x)-1])
    for i in range(1, n):
        res += 2*yf(x[0]+step*i)
    return (x[len(x)-1]-x[0])*res/(2*n)    
```

```python
c2 = trapezd(y, x, 2)
c2, a - c2
```

```python
c4 = trapezd(y, x, 4)
c4, a - c4
```

```python
def simpsn(yf, x, n):
    step = (x[len(x)-1]-x[0])/n
    res = yf(x[0])+yf(x[len(x)-1])
    for i in range(1, n):
        res += (4 if i%2==1 else 2)*yf(x[0]+step*i)
    return (x[len(x)-1]-x[0])*res/(3*n)     
```

```python
e = simpsn(y, x, 5)
e, a - e
```

```python
def simps38(yf, x):
    step = (x[len(x)-1]-x[0])/4
    res = yf(x[0])+yf(x[len(x)-1])+3*yf(x[0]+step)+3*yf(x[0]+step*2)
    return (x[len(x)-1]-x[0])*res/8  
```
