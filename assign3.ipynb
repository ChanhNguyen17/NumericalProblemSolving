### Exercise 02 - Problem 6

```python
# Create variables
maxiter = 10000
k = 1

# Initial f = 0 and calculate the true value of f
f_true = pi**4/90
f = 0

#Iterate k from 1 to maxiter to calculate f
while k <= maxiter:
    f += 1/k**4
    k += 1

# Calculate true error    
eT = (f_true-f)/f_true*100

# Print out the results
print("k iters from 1 to 10000:")
print("Calculated f value is : "+str(f))
print("The true value of f is: "+str(f_true))
print("True error is :         "+str(eT))
print("-------------")

# Reset value of f to 0 and k to maxiter
f = 0
k = maxiter

#Iterate k from maxiter to 1 to calculate f
while k >= 1:
    f += 1/k**4
    k -= 1

# Calculate true error 
eT = (f_true-f)/f_true*100

# Print out the results
print("k iters from 10000 to 1:")
print("Calculated f value is : "+str(f))
print("The true value of f is: "+str(f_true))
print("True error is :         "+str(eT))
print("Done")    
    
```

### Exercise 03 - Problem 1c

```python
# x and y value
x = arange(-2, 2, 0.1)
y = x**5 - x**2 + 2

# plot, label and title
figure()
plot(x, y)
xlabel("x-axis")
ylabel("y-axis")
title("$y = x^5 - x^2 + 2$")
grid()

print()

# Compute the root
iRoot = 0
yClosestToZero = abs(y[0])

# Iterate to find index of the root
for i in range(len(y)):
    if yClosestToZero > abs(y[i]):
        yClosestToZero = abs(y[i])
        iRoot = i

# Plot the root        
plot(x[iRoot], y[iRoot], 'ro')        
print("The root is: "+str(x[iRoot]))

# Compute the local minimum
iMin = 0
iMax = 0
isIncrease = True

# Iterate to find index of the min and max local point
for i in range(len(y)-1):
    if isIncrease:
        if y[i+1] < y[i]:
            iMax = i
            isIncrease = False
    else:
        if y[i+1] > y[i]:
            iMin = i
            isIncrease = True

# Plot the minimum and maximum point            
plot(x[iMin], y[iMin], 'g^')
plot(x[iMax], y[iMax], 'y>')
print("The local minimum point is x = "+str(x[iMin])+" and y = "+str(round(y[iMin], 2)))
print("The local maximum point is x = "+str(x[iMax])+" and y = "+str(round(y[iMax], 2)))
```

### Exercise 03 - Problem 2a

```python
# x and y value
x = arange(-2, 2, 0.01)
y1 = x**3
y2 = 4 - 2*x

# plot, label and title
figure()
plot(x, y1, x, y2)
xlabel("x-axis")
ylabel("y-axis")
title("$y = x^3$ and $y = 4 - 2*x$")
grid()

# Compute the intersection
iInter = 0
yDiffe = y1 - y2
yClosestToZero = abs(yDiffe[0])

# Iterate to find index of the intersection
for i in range(len(x)):
    if yClosestToZero > abs(yDiffe[i]):
        yClosestToZero = abs(yDiffe[i])
        iInter = i

# Plot the intersection        
plot(x[iInter], y1[iInter], 'ro')       
print("The intersection point is x = "+str(x[iInter])+" and y = "+str(round(y1[iInter], 3)))
```
