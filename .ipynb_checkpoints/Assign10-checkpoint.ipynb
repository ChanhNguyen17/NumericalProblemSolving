## Exercise 6
Write my own code that implements the Newton-Raphson method.

```python
def newtonSearch(f, x0, tol, maxiter):
    
    # Initial values
    x = x0
    n = 0
    
    while n < maxiter:
        n += 1
        fx = f(x)
        if fx == 0:
            return x, 0
        dfx = (f(x+tol)-fx)/tol
        try:
            dx = -fx/dfx
        except ZeroDivisionError:
            dx = x0
        x += dx
        
        # Stop criteria
        ea = abs(fx)
        if ea < tol:
            return x, n
        
    return x, n
```

```python
x = np.linspace(-2, 2, 1000)
def f1(x):
    return x**2 - x - 1
def f2(x):
    return x**3 - 3*sin(x)
def f3(x):
    return exp(x) - 2
def f4(x):
    return 1 - x**2 + sin(50/(1+x**2))

fig, axes = plt.subplots(1, 4, figsize=(12, 3), sharey=True)
 
for n, y in enumerate([f1(x), f2(x), f3(x), f4(x)]):
    axes[n].plot(x, y)
    axes[n].grid()
    axes[n].set_xticks([-2, -1, 0, 1, 2])
    axes[n].axhline(0, color = 'r')
```

```python
# Find the root of f1(x) with x0 = 0.0
x0, n = newtonSearch(f1, 0.0, 0.01, 1000)
print('x0, n =', (x0, n))
```

Determine the real root of  $f(x)=-26+85x-91x^2+44x^3-8x^4+x^5$

###### By using graphically method, the root is x = 0.5570

```python
def fx(x0):
    return 26/x0**4-85/x0**3+91/x0**2-44/x0+8

def simpleFixedPointIteration(x0, fx, eS, maxIter):
    ite = 0
    xr = x0
    ea = 1.0
    while(ite<maxIter):
        ite+=1
        xrold = xr
        xr = fx(xr)
        ea = abs((xr-xrold)/xr)*100
        if(ea<eS):
            return xr, ite
    return xr, ite
```

###### (b) Using simple fixed-point iteration

```python
x, n = simpleFixedPointIteration(0.5, fx, 0.01, 100000)
print('x, n =', (x, n))
```

###### (b) Find the roots using the fixed-point iteration

```python
def fx(x0):
    return (x0[0]**2-x0[1])/(5*x0[1])
def fy(x0):
    return -x0[0]**2+x0[0]+0.75

def fixedPointIteration(x0, fx, fy, maxIter):
    xr = x0
    ite = 0
    while(ite<maxIter):
        ite+=1
        xrold = xr
        xr[0] = fx(x0)
        xr[1] = fy(x0)
    return xr
```

```python
fixedPointIteration([-0.5,-0.5], fx, fy, 5000)
```

```python
fixedPointIteration([1.37,0.23], fx, fy, 5000)
```

```python
fixedPointIteration([-0.59,-0.18], fx, fy, 5000)
```
