## Exercise 2
Write a function for incremental search and test it with the functions given in previous exercise

```python
fig,axes = plt.subplots(1,4,figsize=(12,3),sharey=True)
axes[0].plot(x,f1(x))
axes[1].plot(x,f2(x))
axes[2].plot(x,f3(x))
axes[3].plot(x,f4(x))

for n in range(4):
    axes[n].grid()
    axes[n].set_xticks([-2,-1,0,1,2])
    axes[n].axhline(0,color='r')
```

```python
def f1(x):
    return x**2 - x - 1

def incrementSearch(f,a,b,dx):
    x0 = a
    n = 0
    while x0 < b:
        if sign(f(x0)) != sign(f(x0+dx)):
            return x0, n
        x0 += dx
        n += 1
    return None, n
```

```python
incrementSearch(f1,-2,2,0.01)
```
