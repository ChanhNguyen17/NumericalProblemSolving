```python
%pylab inline
from scipy.stats import linregress
```

```python
x=array([0,2,4,6,9,11,12,15,17,19])
y=array([5,6,7,6,9,8,7,10,12,12])
plot(x,y,'o')
xlim(-1,21)
ylim(0,15)
grid()

r = linregress(x,y)
a1 = r.slope
a0 = r.intercept
x2 = linspace(0,20)
y2 = a0+a1*x2
plot(x2,y2,'r')
```

```python
results = linregress(x,y)
results
```

```python
x = array([2.5, 3.5, 5, 6, 7.5, 10, 12.5, 15, 17.5, 20])
y = array([13, 11, 8.5, 8.2, 7, 6.2, 5.2, 4.8, 4.6, 4.3])
plot(x, y, 'o')
grid()

r = linregress(x, y)
x2 = linspace(1, 22)
y2 = r.slope*x2 + r.intercept

# a = exp(r.slope)
# b = r.intercept
# y3 = a*x2**b
# plot(x2, y22, 'r')

plot(x2, y2)
```

```python
xlog = log(x)
ylog = log(y)
plot(xlog, ylog, 'o')
grid()

r = linregress(xlog, ylog)
x3 = linspace(0.5, 3.5, 10)
y3 = r.slope*x3 + r.intercept
plot(x3, y3)
```

```python
plot(x, y, 'o')

xlog = log(x)
ylog = log(y)
r = linregress(xlog, ylog)

a2 = r.intercept
b2 = r.slope
a = exp(a2)
b = b2
x2 = linspace(2, 20)
y2 = a*x2**b
plot(x2, y2)
ylim(4, 14)
grid()
```

```python
# Curve fitting way
from scipy.optimize import curve_fit

def powerfun(x, a, b):
    return a*x**b

x2 = linspace(2, 20)
p, pcov = curve_fit(powerfun, x, y, p0 = (20, -1.0))
a = p[0]
b = p[1]
p
```

```python
ax = linspace(2, 20, 100)
a = p[0]
b = p[1]
y2 = powerfun(x2, a, b)
```
