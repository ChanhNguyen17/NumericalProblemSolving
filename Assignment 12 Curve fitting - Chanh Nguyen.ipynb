## Exercise 3
Fit an exponential model to   
<table style="width:100%">
    <tr>
        <th>x</th>
        <td>0.4</td>
        <td>0.8</td>
        <td>1.2</td>
        <td>1.6</td>
        <td>2.0</td>
        <td>2.3</td>
    </tr>
    <tr>
        <th>y</th>
        <td>800</td>
        <td>975</td>
        <td>1500</td>
        <td>1950</td>
        <td>2900</td>
        <td>3600</td>
    </tr>
</table>  
    
Plot the data and the equation on both standard and semi-logarithmic graph.   


```python
x=array([0.4, 0.8, 1.2, 1.6, 2, 2.3])
y=array([800, 975, 1500, 1950, 2900, 3600])

fig, ax = subplots(1, 3, figsize = (12, 3))

# The data has a linear relationship between x and log(y). Use linregress for x and log(y)
r = linregress(x, log(y))
a = r.slope
b = r.intercept

# Create equation based on linregress result
x2 = linspace(0, 2.5)
# The exponential model between x and y
y2 = exp(a*x2 + b)

# Plot the data and equation on standard graph
sca(ax[0])
plot(x, y, 'o')
plot(x2,y2)
title('linear plot')
grid()

# Plot the data and equation on semi-log(x)
sca(ax[1])
semilogx(x, y, 'o')
semilogx(x2, y2)
title('semilogx plot')
grid()

# Plot the data and equation on semi-log(y)
sca(ax[2])
semilogy(x, y, 'o')
semilogy(x2, y2)
title('semilogy plot')
grid()

'The model equation: y = exp(a*x + b) with a = '+str(r.slope)+' and b = '+str(r.intercept)
```

## Exercise 5
Use nonlinear regression to fit a parabola to the following data

<table style="width:100%">
    <tr>
        <th>x</th>
        <td>0.2</td>
        <td>0.5</td>
        <td>0.8</td>
        <td>1.2</td>
        <td>1.7</td>
        <td>2.0</td>
        <td>2.3</td>
    </tr>
    <tr>
        <th>y</th>
        <td>500</td>
        <td>700</td>
        <td>1000</td>
        <td>1200</td>
        <td>2200</td>
        <td>2650</td>
        <td>3750</td>
    </tr>
</table> 

```python
x = array([0.2, 0.5, 0.8, 1.2, 1.7, 2, 2.3])
y = array([500, 700, 1000, 1200, 2200, 2650, 3750])

fig, ax = subplots(1, 2, figsize = (12, 5))

# The data has a linear relationship between x and log(y). Use linregress for x and log(y)
ylog = log(y)
r = linregress(x, ylog)
a = r.slope
b = r.intercept

# Create equation based on linregress result
x2 = linspace(0, 2.5)
y2log = a*x2 + b

# Plot graph x and log(y)
sca(ax[1])
title('x and log(y)')
plot(x, ylog, 'o')
plot(x2, y2log)
grid()

# Plot graph x and y
sca(ax[0])
y2 = exp(y2log)
plot(x, y, 'o')
plot(x2, y2)
title('x and y')
grid()

'The model equation: y = exp(a*x + b) with a = '+str(r.slope)+' and b = '+str(r.intercept)
```

## Exercise 6
A material is tested for cyclic fatique failure whereby a stress, in MPa,
is applied to the material and the number of cycles needed to cause
failure is measured. The results are in the table below. When a loglog
plot of stress versus cycles is generated, the data trend shows a
linear relationship. Use least-squares regression to determing a best-fit
equation for this data.

<table style="width:100%">
    <tr>
        <th>N, cycles</th>
        <td>1</td>
        <td>10</td>
        <td>100</td>
        <td>1000</td>
        <td>10000</td>
        <td>100000</td>
        <td>1000000</td>
    </tr>
    <tr>
        <th>Stress, MPa</th>
        <td>1100</td>
        <td>1000</td>
        <td>925</td>
        <td>800</td>
        <td>625</td>
        <td>550</td>
        <td>420</td>
    </tr>
</table> 

```python
x = array([1, 10, 100, 1000, 10000, 100000, 1000000])
y = array([1100, 1000, 925, 800, 625, 550, 420])

fig, ax = subplots(1, 2, figsize = (12, 5))

# The data has a linear relationship between log(x) and log(y). Use linregress for log(x) and log(y)
xlog = log(x)
ylog = log(y)
r = linregress(xlog, ylog)
a = r.slope
b = r.intercept

# Create equation based on linregress result
x2 = linspace(1, 1000000, 1000000)
x2log = log(x2)
y2log = a*x2log + b

# Plot graph log(x) and log(y)
sca(ax[1])
title('log(x) and log(y)')
plot(xlog, ylog, 'o')
plot(x2log, y2log)
grid()

# Plot graph x and y
sca(ax[0])
y2 = exp(y2log) # Best fit equation y = exp(a*log(x) + b)
plot(x, y, 'o')
plot(x2, y2)
title('x and y')
grid()

'The model equation: y = exp(a*log(x) + b) with a = '+str(r.slope)+' and b = '+str(r.intercept)
```
